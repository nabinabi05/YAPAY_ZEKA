```python
# ========== HÜCRE 2: LLVIP'i extract et ==========
import os, zipfile

zip_path    = '/content/drive/MyDrive/LLVIP.zip'
extract_to  = '/content/drive/MyDrive/'
check_path  = '/content/drive/MyDrive/LLVIP'

if os.path.exists(check_path):
    print('✓ LLVIP zaten extract edilmiş, atlanıyor')
else:
    print('Extracting LLVIP.zip (birkaç dakika sürebilir)...')
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
    print('✓ LLVIP extract tamamlandı')
```
